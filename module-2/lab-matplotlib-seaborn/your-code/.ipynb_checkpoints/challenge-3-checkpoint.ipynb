#### Using Seaborn plot the correlation matrix of various features.

*Hint: search how to use `heatmap`*

```python
#your code here
corr =  titanic.corr()
mask = np.triu(np.ones_like(corr, dtype = np.bool))
sns.heatmap(corr,linewidths=.5,cmap='YlGnBu', mask=mask,square=True);
```
